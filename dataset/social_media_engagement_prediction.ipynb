Removing Outliers Using IQR Method
</h3>

</div>


```python
# identify outliers using IQR on all numeric columns

# select numeric columns
num_cols = df.select_dtypes(include=np.number).columns

# compute Q1, Q3 and IQR
Q1 = df[num_cols].quantile(0.25)
Q3 = df[num_cols].quantile(0.75)
IQR = Q3 - Q1

# boolean dataframe marking where values are outside the 1.5·IQR range
outlier_mask = ((df[num_cols] < (Q1 - 1.5 * IQR)) |
                (df[num_cols] > (Q3 + 1.5 * IQR)))

# rows that contain at least one outlier
outlier_rows = df[outlier_mask.any(axis=1)]

#remove the outliers
df_clean = df[~outlier_mask.any(axis=1)]

print("original rows: " ,df.shape)
print("after removing outliers: " , df_clean.shape)

```

```python
# Encode categorical variables using LabelEncoder
categorical_cols = df_clean.select_dtypes(include='object').columns

for col in categorical_cols:
    le = LabelEncoder()
    df_clean[col] = le.fit_transform(df_clean[col])

print("Categorical variables encoded successfully")
```
